```python
import requests
import json
import time  #library used to get current time
import pandas as pd
```

```python
#create a text file that contains your API key
#you DO NOT want others to see your API key
#this method is not the most secure but will work for this exercise

#this will read the file and store your API key as a variable
with open('book_key.txt', 'r') as file:
    for line in file:
        api_key = line
```

```python
api_key
```

```python
url=r"https://api.nytimes.com/svc/books/v3/lists/current/hardcover-fiction.json?api-key="
```

```python
response = requests.get(url+api_key)
```

```python
response
```

```python
response.text
```

```python
#set the string data to a variable
str_book = response.text
```

```python
#load in the string data as JSON
book = json.loads(str_book)
```

```python
PrimaryISBN10=[]
PrimaryISBN13=[]
Title=[]
Author=[]
Rank=[]
LastWeeksRank=[]
NumberOfWeeks=[]
Price=[]
Description=[]
Publisher=[]
BookImage=[]
```

```python
PrimaryISBN10
```

```python
for nyt in book['results']['books']:
    PrimaryISBN10.append(nyt['primary_isbn10'])
    PrimaryISBN13.append(nyt['primary_isbn13'])
    Title.append(nyt['title'])
    Author.append(nyt['author'])
    Rank.append(nyt['rank'])
    LastWeeksRank.append(nyt['rank_last_week'])
    NumberOfWeeks.append(nyt['weeks_on_list'])
    Price.append(nyt['price'])
    Description.append(nyt['description'])
    Publisher.append(nyt['publisher'])
    BookImage.append(nyt['book_image'])
```

```python
print(len(PrimaryISBN10))
print(len(PrimaryISBN13))
print(len(Title))
print(len(Author))
print(len(Rank))
print(len(LastWeeksRank))
print(len(NumberOfWeeks))
print(len(Price))
print(len(Description))
print(len(Publisher))
print(len(BookImage))
```

```python
table = list(zip(PrimaryISBN10,PrimaryISBN13,Title,Author,Rank,LastWeeksRank,NumberOfWeeks,
                                Price,Description,Publisher,BookImage))
```

```python
columnName = ['primary ISBN10','primary ISBN13','title','author','rank',"last week's rank",'number of weeks on the list',
          'price','description','publisher','book image']
```

```python
df = pd.DataFrame(table,columns = columnName)
df
```
